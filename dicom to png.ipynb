```python
import os
import cv2
import pydicom

def convert_dcm_to_png(dcm_folder, png_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(png_folder):
        os.makedirs(png_folder)

    # Get a list of all DICOM files in the input folder
    dcm_files = [f for f in os.listdir(dcm_folder) if f.endswith('.dcm')]

    # Loop over each DICOM file
    for dcm_file in dcm_files:
        # Read the DICOM file
        dcm_path = os.path.join(dcm_folder, dcm_file)
        ds = pydicom.dcmread(dcm_path)

        # Get the pixel data as a NumPy array
        pixel_array = ds.pixel_array

        # Convert to 8-bit unsigned integer (0-255)
        pixel_array = pixel_array.astype('uint8')

        # Save as PNG image
        png_file = dcm_file[:-4] + '.png'
        png_path = os.path.join(png_folder, png_file)
        cv2.imwrite(png_path, pixel_array)

        print(f"Converted {dcm_file} to {png_file}")

# Specify the input DICOM folder and the output PNG folder
dcm_folder = 'D:/Tez data/data/chaos/Train_Sets/CT/2/DICOM_anon'
png_folder = 'D:/Tez data/data/chaos/Train_Sets/CT/2/Images'

# Call the conversion function
convert_dcm_to_png(dcm_folder, png_folder)
```
